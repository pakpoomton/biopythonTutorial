```python
# reading sequence file

from Bio import SeqIO

for seq_record in SeqIO.parse("ls_orchid.fa", "fasta"):
    print(seq_record.id)
    print(repr(seq_record.seq))
    print(len(seq_record))
    

for seq_record in SeqIO.parse("ls_orchid.gb", "genbank"):
    print(seq_record.id)
    print(repr(seq_record.seq))
    print(len(seq_record))
    

# parse file, get each seq_record and put all id into a list    
identifiers = [seq_record.id for seq_record in SeqIO.parse("ls_orchid.gb", "genbank")]
print(identifiers)




```

```python
# iterating over recorded sequence file
from Bio import SeqIO

record_iterator = SeqIO.parse("ls_orchid.fa", "fasta")

first_record = next(record_iterator)
print(first_record.id)
print(first_record.description)

second_record = next(record_iterator)
print(second_record.id)
print(second_record.description)

```

```python
##***** getting a list of recorded sequence ***

from Bio import SeqIO
records = list(SeqIO.parse("ls_orchid.gb", "genbank"))
print("Found %i records" % len(records))

print("The last record")
last_record = records[-1]  # using Python's list tricks
print(last_record.id)
print(repr(last_record.seq))
print(len(last_record))

print("The first record")
first_record = records[0]  # remember, Python counts from zero
print(first_record.id)
print(repr(first_record.seq))
print(len(first_record))
```

```python
### ** extracting data

## for loop extract
all_species = []
for seq_record in SeqIO.parse("ls_orchid.gb", "genbank"):
    all_species.append(seq_record.annotations["organism"]) # extract species names
print(all_species)

# list comprehension extract
all_species = [
    seq_record.annotations["organism"]
    for seq_record in SeqIO.parse("ls_orchid.gb", "genbank")
]
print(all_species)
```

```python
# extract data from fasta file

# using for loop
all_species = []
for seq_record in SeqIO.parse("ls_orchid.fa", "fasta"):
    all_species.append(seq_record.description.split()[1]) ## second item on the the header of fasta ~ 
print(all_species)

# list comprehension extract
all_species = [
    seq_record.description.split()[1]
    for seq_record in SeqIO.parse("ls_orchid.fa", "fasta")
]
print(all_species)
```

```python
### parsing sequences from the net, fasta

from Bio import Entrez
from Bio import SeqIO

Entrez.email = "A.N.Other@example.com"
with Entrez.efetch(
    db="nucleotide", rettype="fasta", retmode="text", id="6273291"
) as handle:
    seq_record = SeqIO.read(handle, "fasta")
print("%s with %i features" % (seq_record.id, len(seq_record.features)))
```

```python
### parsing sequences from the net, genbank

from Bio import SeqIO

Entrez.email = "A.N.Other@example.com"
with Entrez.efetch(
    db="nucleotide", rettype="gb", retmode="text", id="6273291"
) as handle:
    seq_record = SeqIO.read(handle, "gb")  # using "gb" as an alias for "genbank"
print("%s with %i features" % (seq_record.id, len(seq_record.features)))
```

```python
## accessing multiple records
from Bio import Entrez
from Bio import SeqIO

Entrez.email = "A.N.Other@example.com"
with Entrez.efetch(
    db="nucleotide", rettype="gb", retmode="text", id="6273291,6273290,6273289"
) as handle:
    for seq_record in SeqIO.parse(handle, "gb"):
        print("%s %s..." % (seq_record.id, seq_record.description[:50]))
        print(
            "Sequence length %i, %i features, from: %s"
            % (
                len(seq_record),
                len(seq_record.features),
                seq_record.annotations["source"],
            )
        )
```

```python
## parsing sequences from the net, swissprot 

from Bio import ExPASy
from Bio import SeqIO

with ExPASy.get_sprot_raw("O23729") as handle:
    seq_record = SeqIO.read(handle, "swiss")
print(seq_record.id)
print(seq_record.name)
print(seq_record.description)
print(repr(seq_record.seq))
print("Length %i" % len(seq_record))
print(seq_record.annotations["keywords"])
```

```python
# sequence file as dictionary

from Bio import SeqIO
orchid_dict = SeqIO.to_dict(SeqIO.parse("ls_orchid.gb", "genbank"))

print(len(orchid_dict))
print(list(orchid_dict.keys()))

seq_record = orchid_dict["Z78475.1"]
print("\n")
print(seq_record.description)
```
